```python
# Importing The Libraries
```

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

```python
# Reading the Orignal Image
```

```python
im=cv2.imread("C:/Users/Rushikesh/Desktop/yash.jpg")
```

```python
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(im))
```

```python
# Converting the Original Image into Grayscale
```

```python
gray_image = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

```python
# Inverting the Grayscale image converted previously 
```

```python
inv_gray_image = 255 - gray_image
```

```python
inv_gray_image
```

```python
display(Image.fromarray(inv_gray_image))
```

```python
# Making the image Blur 
```

```python
blur_image = cv2.GaussianBlur(inv_gray_image, (35, 35), 0)
```

```python
display(Image.fromarray(blur_image))
```

```python
# Mixing the grayscale image and inverted blur image
```

```python
def dodge(gray_image, blur_image):
    inv_blur_img = 255 - blur_image
    return cv2.divide(gray_image, inv_blur_img, scale = 450.0)
inv_blur_img = 255 - blur_image
```

```python
# Enhanced image
```

```python
pencil_sketch = dodge(gray_image, inv_blur_img)
display(Image.fromarray(pencil_sketch))
```
